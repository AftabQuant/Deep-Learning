```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_moons
```

```python
x, y = make_moons(n_samples=300, noise=0.2, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
import tensorflow
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann = Sequential()
```

```python
ann.add(Dense(10, activation='sigmoid', input_dim = 2))
ann.add(Dense(10, activation='sigmoid'))
ann.add(Dense(10, activation='sigmoid'))
ann.add(Dense(10, activation='sigmoid'))
ann.add(Dense(10, activation='sigmoid'))
ann.add(Dense(10, activation='sigmoid'))
ann.add(Dense(10, activation='sigmoid'))
ann.add(Dense(1, activation='sigmoid'))
```

```python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
old_weights = ann.get_weights()[0]
```

```python
old_weights
```

```python
ann.fit(x_train, y_train, epochs=1)
```

```python
new_weights = ann.get_weights()[0]
new_weights
```

```python
gradients = (new_weights - old_weights) / 0.001
percent_change = abs(gradients) * 100
```
